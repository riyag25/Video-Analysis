## Function to get final timestamps

```python
# add buffer time 50 seconds before and after and output the final video timestamp
def get_sec(time):
  l = time.split(':')
  min = int(l[-2])
  sec= int(l[-1])
  if len(l)==3:
    hr = int(l[0])
    time = hr*60*60 + min *60 + sec
  time =  min *60 + sec # if no hr then only min and sec
  return (time-30, time+30)
```

```python
import json
# Opening JSON file
def final_video(path):
  """ Function takes path of the json file, extract goal events, which has confidence >25%, 
  convert time of the event to seconds, add 30sec buffer and returns a list of tuples with (start,end)
  for each goal in the video.
  """
  f = open(path)
  # returns JSON object as # a dictionary
  data = json.load(f)
  # get time and confidence for the label goal 
  time = []
  for frame in data['predictions']:
    if (frame['label']=='Goal'):
      time.append((frame['gameTime'], frame['confidence']))
  # get the goal time if confidence for label goal is >25%
  goals = [t[0] for t in time if float(t[1])>.25]   
  #exact time of goal predicted
  gtime = [g.split()[-1] for g in goals]
  # add buffer time 50 seconds before and after and output the final video timestamp
  video_time = [get_sec(time) for time in gtime]
  return video_time
```

```python
final_video('inference/outputs/Predictions-v2.json')
```

## Store timestamped videos

```python
predictions = final_video('inference/outputs/Predictions-v2.json') # get start and end for goal predictions
print(f"Number of goals in this game is {len(predictions)}") 
```

### Timestamped video saving

```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
start=[i[0] for i in predictions] #get start and end for each clip
end=[i[1] for i in predictions]
for i,t in enumerate(start):
     
  filename="inference/outputs/highlight" + str(i+1) + ".mp4"
  ffmpeg_extract_subclip("1_224p.mkv",start[i],end[i],targetname=filename)
```
